# Tổng quan về Phân tích và xử lý dữ liệu trong Python (Pandas, Numpy, MatPlotLib)

### Pandas DataFrames và Visualization Basics

Biết những thư viện nào cần cài đặt và các quy ước thích hợp là rất quan trọng. Sử dụng np cho NumPy, pd cho Pandas và plt cho PyPlot (MatPlotLib). Đây là những thư viện cơ bản phục vụ việc phân tích và xử lý dữ liệu đó.

```python
# important imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Hãy bắt đầu bằng cách tạo và kiểm tra xem bộ dữ liệu DataFrame(df) gồm kiểu dữ liệu nào

```python
names = ['John', 'Mary', 'Kelly', 'Nicole', 'Tony', 'Chris']
ages = [26, 28, 18, 24, 19, 20]
testScores = [100, 95, 60, 90, 50.5, 70]

data = list(zip(names, ages, testScores))
data
```

```python
df = pd.DataFrame(data, columns= ['Names', 'Ages', 'Scores'])
df
```

```python
df.dtypes
```

```python
df.shape
```

Dưới đây là một số hàm cơ bản mà các bạn có thể dùng để hiểu rõ hơn về bộ dữ liệu mà các bạn đang có

```python
sortedAge = df.sort_values(['Ages'], ascending=True)
sortedScore = df.sort_values(['Scores'], ascending=False)
```

```python
sortedAge
```

```python
sortedScore
```

```python
df['Names']
```

Giờ thì hãy xem cách chúng ta có thể trực quan hóa bộ dữ liệu nhé!

```python
df['Ages'].plot()
```

```python
plt.figure()
plt.plot(df['Ages'], df['Scores'], 'ro')
plt.xlabel('Age')
plt.ylabel('Test Score')
plt.title('Ages vs Test Scores')
plt.show()
```

```python
plt.figure(figsize=(9, 3))
plt.bar(df['Names'], df['Scores'])
plt.xlabel('Student')
plt.ylabel('Test Score')
plt.title('Student Test Scores')
plt.show()
```

### Cách tải bộ dữ liệu có dạng '.cvs' và thao tác với dữ liệu

```python
data_url = "https://raw.githubusercontent.com/arjun-prabhakar/crimson-data-science/main/boston.csv"
bostonData = pd.read_csv(data_url)
print(bostonData.columns)
```

```python
bostonData.head()
```

```python
len(bostonData)
```

Bằng cách sử dụng tập dữ liệu được chọn, chúng ta có thể kiểm tra các chức năng khác nhau mà thư viện Pandas cho phép chúng ta sử dụng trên các thành phần của bộ dữ liệu DataFrame

```python
values = ['age', 'tax', 'medv']

df = bostonData[values]
print(df['age'][0])

df.head()
```

```python
print('Maximum Age: {}'.format(df['age'].max()))
print('Minimum Age: {}'.format(df['age'].min()))
```

```python
df.describe()
```

```python
#loc gets rows (and/or columns) with certain labels
df.loc[4]
```

```python
#iloc gets rows (and/or columns) at integer locations
df.iloc[0]
```
